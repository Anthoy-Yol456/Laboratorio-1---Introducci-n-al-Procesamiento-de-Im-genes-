## Problema 1:

```python
def modify_image(image_path, color):
    image = cv2.imread(image_path)
    
    # Mascaras
    mask = np.zeros_like(image)
    
    # Diccionario para los colores
    channels = {
        1: [0],       # Solo azul
        2: [1],       # Solo verde
        3: [2],       # Solo rojo
        10: [1, 2],   # Verde y rojo
        20: [0, 1],   # Azul y verde
        30: [0, 2]    # Azul y rojo
    }
    
    if color in channels:
        for ch in channels[color]:
            mask[:, :, ch] = image[:, :, ch]
    
    return mask
    
```

```python
image_path = 'tikal.jpg' # Aqui se carga la imagen
color = 3  # Cambia este valor para probar diferentes casos
modified_image = modify_image(image_path, color)

original_image = cv2.imread(image_path)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Imagen Modificada')
plt.imshow(cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB))

plt.show()
```

## Problema 3

```python
def crear_escala_grises_3d(imagen_path):
    imagen = cv2.imread(imagen_path)

    imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    
    imagen_3d_1 = np.zeros_like(imagen)
    imagen_3d_2 = np.zeros_like(imagen)
    imagen_3d_3 = np.zeros_like(imagen)
    
    imagen_3d_1[:,:,0] = imagen_gris
    imagen_3d_2[:,:,1] = imagen_gris
    imagen_3d_3[:,:,2] = imagen_gris
    
    # Mostrar las imágenes
    plt.figure(figsize=(10, 10))
    
    plt.subplot(1, 3, 1)
    plt.title('Imagen Gris en Dim 1')
    plt.imshow(cv2.cvtColor(imagen_3d_1, cv2.COLOR_BGR2RGB))
    
    plt.subplot(1, 3, 2)
    plt.title('Imagen Gris en Dim 2')
    plt.imshow(cv2.cvtColor(imagen_3d_2, cv2.COLOR_BGR2RGB))
    
    plt.subplot(1, 3, 3)
    plt.title('Imagen Gris en Dim 3')
    plt.imshow(cv2.cvtColor(imagen_3d_3, cv2.COLOR_BGR2RGB))
    
    plt.show()
    
    return imagen_3d_1, imagen_3d_2, imagen_3d_3

imagen_path = 'quetzal.jpg'
crear_escala_grises_3d(imagen_path)
```

## Problema 4

```python
def calcular_histograma(imagen, canal, color, titulo):
    hist, bins = np.histogram(imagen[:, :, canal].ravel(), 256, [0, 256])
    media = np.mean(imagen[:, :, canal])
    
    plt.plot(hist, color=color)
    plt.axvline(x=media, color='k', linestyle='--', linewidth=1)
    plt.title(titulo)
    plt.xlabel('Intensidad de píxeles')
    plt.ylabel('Número de píxeles')
    plt.show()
```

```python
def calcular_histograma_gris(imagen_gris):
    hist_gris, bins_gris = np.histogram(imagen_gris.ravel(), 256, [0, 256])
    media_gris = np.mean(imagen_gris)
    
    plt.plot(hist_gris, color='gray')
    plt.axvline(x=media_gris, color='k', linestyle='--', linewidth=1)
    plt.title('Histograma de la Imagen en Escala de Grises')
    plt.xlabel('Intensidad de píxeles')
    plt.ylabel('Número de píxeles')
    plt.show()
```

```python
def mostrar_histograma_azul(imagen):
    calcular_histograma(imagen, 0, 'b', 'Histograma del Canal Azul')

def mostrar_histograma_verde(imagen):
    calcular_histograma(imagen, 1, 'g', 'Histograma del Canal Verde')

def mostrar_histograma_rojo(imagen):
    calcular_histograma(imagen, 2, 'r', 'Histograma del Canal Rojo')

def mostrar_histograma_gris(imagen):
    imagen_gris = np.mean(imagen, axis=2).astype(np.uint8)
    calcular_histograma_gris(imagen_gris)
```

```python
imagen_path = 'quetzal.jpg'
imagen = cv2.imread(imagen_path)
```

```python
mostrar_histograma_azul(imagen)
mostrar_histograma_verde(imagen)
mostrar_histograma_rojo(imagen)
mostrar_histograma_gris(imagen)
```

## Problema 5

```python
def weighted_grayscale(image_path):
    image = cv2.imread(image_path)

    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    R, G, B = image_rgb[:,:,0], image_rgb[:,:,1], image_rgb[:,:,2]
    gray_image = 0.299 * R + 0.587 * G + 0.114 * B
    gray_image = gray_image.astype(np.uint8)
    
    return gray_image
```

```python
def display_images(original_image_path, gray_image):
    original_image = cv2.imread(original_image_path)
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.title('Imagen Original')
    plt.imshow(original_image_rgb)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Imagen en Escala de Grises Ponderada')
    plt.imshow(gray_image, cmap='gray')
    plt.axis('off')

    plt.show()
```

```python
image_path = 'quetzal.jpg'
gray_image = weighted_grayscale(image_path)
display_images(image_path, gray_image)
```

## Problema 6

El espacio de color HSV (Hue, Saturation, Value) es una representación de colores que se basa en tres componentes:

1. Hue (Tono): Representa el color puro y se mide en grados, de 0 a 360. Comienza en el rojo (0°), pasa por el verde (120°), azul (240°) y vuelve al rojo.

2. Saturation (Saturación): Indica la pureza o intensidad del color, variando de 0% a 100%. Un 0% es un color gris sin saturación (blanco o negro) y un 100% es el color completamente saturado.

3. Value (Valor): Describe el brillo del color, también de 0% a 100%. Un 0% es completamente oscuro (negro) y un 100% es el color con el máximo brillo.

Para mapear colores al espacio HSV desde el espacio RGB (Red, Green, Blue), se sigue un procedimiento matemático:

1. Normalizar los valores RGB a un rango de 0 a 1.
2. Encontrar el valor máximo (max) y mínimo (min) de los valores normalizados.
3. Calcular el Value (V) como el valor máximo.
4. Calcular la Saturation (S) usando la fórmula: 
   S = 
   - 0 si V = 0
   - (V - min) / V si V ≠ 0
5. Calcular el Hue (H) dependiendo de cuál de los valores RGB es el máximo y usar fórmulas específicas para cada caso.

Ejemplo Sencillo

Supongamos que tenemos un color RGB con valores (255, 0, 0), que es el color rojo puro.

1. Normalizamos los valores RGB a un rango de 0 a 1:
   - R = 255 / 255 = 1
   - G = 0 / 255 = 0
   - B = 0 / 255 = 0

2. Encontramos los valores máximo y mínimo:
   - max = 1
   - min = 0

3. Calculamos el Value (V):
   - V = max = 1

4. Calculamos la Saturation (S):
   - S = (max - min) / max = (1 - 0) / 1 = 1 (100%)

5. Calculamos el Hue (H):
   - Dado que max es R, usamos la fórmula:
     H = 60 * ((G - B) / (max - min)) = 60 * ((0 - 0) / (1 - 0)) = 0°

Así, el color rojo puro (255, 0, 0) en el espacio HSV se representa como (0°, 100%, 100%).
